## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Add a new column to insert the age group each player belongs to
heroesdf["Age Groups"] = pd.cut(heroesdf["Age"],age_bins, labels=group_names)

#Filter by Age group and player name
AgeAnalysis = heroesdf.groupby("Age Groups")["SN"].nunique().reset_index()

#Find the Percentage and add to heroesdf
AgeAnalysis["Percentage of Player Age"]=round(100*AgeAnalysis["SN"]/PlayerCount,2)

#Change column Name,sort and remove index
SummaryAge=AgeAnalysis.rename(columns = {"SN": "Total Age Count"}).set_index("Age Groups")
SummaryAge["Percentage of Player Age"]=SummaryAge["Percentage of Player Age"].map("{:,.2f}%".format)
SummaryAge
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Group by Gender and Price and going to use the aggregate function
AgePurchasingAnalysis = round(heroesdf.groupby("Age Groups")["Price"].agg(["sum", "mean","count"]),2)
AgePurchasingAnalysis=AgePurchasingAnalysis.rename(columns = {"sum": "Total Purchase Value Age", "mean": "Average Purchase Price Age", 
                                                                    "count":"Purchase Count Age" })

#Add the calculations to the dataframe
AgePurchasingAnalysis["Avg Purchase Price per Person"]=round(AgePurchasingAnalysis["Total Purchase Value Age"]/SummaryAge["Total Age Count"], 2)


#forma with dollar sign
AgePurchasingAnalysis["Total Purchase Value Age"]=AgePurchasingAnalysis["Total Purchase Value Age"].map("${:,.2f}".format)
AgePurchasingAnalysis["Average Purchase Price Age"]=AgePurchasingAnalysis["Average Purchase Price Age"].map("${:,.2f}".format)
AgePurchasingAnalysis["Avg Purchase Price per Person"]=AgePurchasingAnalysis["Avg Purchase Price per Person"].map("${:,.2f}".format)
AgePurchasingAnalysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by going to use the aggregate function
TopSpender= round(heroesdf.groupby("SN")["Price"].agg(["sum", "mean","count"]),2)

TopSpenderAnalysis=TopSpender.rename(columns = {"sum": "Total Purchase Value", "mean": "Average Purchase Price", 
                                                                   "count":"Purchase Count" })
TopSpenderAnalysis=TopSpenderAnalysis[["Purchase Count","Average Purchase Price","Total Purchase Value"]].sort_values(["Purchase Count"],ascending = False)

TopSpenderAnalysis["Average Purchase Price"]=TopSpenderAnalysis["Average Purchase Price"].map("${:,.2f}".format)
TopSpenderAnalysis["Total Purchase Value"]=TopSpenderAnalysis["Total Purchase Value"].map("${:,.2f}".format)

Top5Spenders=(TopSpenderAnalysis[:5])
Top5Spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
PopularItems= round(heroesdf.groupby(["Item ID","Item Name"])["Price"].agg(["count","sum","mean"]),2)

PopularItemsexc=PopularItems.drop(["mean"], axis=1)

TopPopularItems=PopularItemsexc.rename(columns = {"count": "Total Purchase Count", "sum": "Total Purchase Price"})
TopPopularItems=TopPopularItems[["Total Purchase Count","Total Purchase Price"]].sort_values(["Total Purchase Count"],ascending = False)

TopPopularItems["Total Purchase Price"]=TopPopularItems["Total Purchase Price"].map("${:,.2f}".format)


Top5PopularItems=(TopPopularItems[:5])
Top5PopularItems
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
TopPopularAll=PopularItems.rename(columns = {"count": "Total Purchase Count", "mean": "Item Price","sum": "Total Purchase Price"})
TopPopularAll=TopPopularAll[["Total Purchase Count","Item Price","Total Purchase Price" ]].sort_values(["Total Purchase Count"],ascending = False)

TopPopularAll["Item Price"]=TopPopularAll["Item Price"].map("${:,.2f}".format)
TopPopularAll["Total Purchase Price"]=TopPopularAll["Total Purchase Price"].map("${:,.2f}".format)


Top5PopularAll=(TopPopularAll[:5])
Top5PopularAll
```
